```python
import pandas as pd
from langchain_community.llms import HuggingFacePipeline
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.chains import ConversationalRetrievalChain
from langchain.memory import ConversationBufferMemory
from transformers import pipeline
```

```python
# Load Hugging Face pipeline (free model)
generator = pipeline("text2text-generation", model="google/flan-t5-small", max_length=256)

# Wrap it with LangChain
llm = HuggingFacePipeline(pipeline=generator)
```

```python
# Upload your own CSV from local machine
from google.colab import files
uploaded = files.upload()

# Load CSV (replace filename with your uploaded file name)
csv_file = list(uploaded.keys())[0]
df = pd.read_csv(csv_file)

print("Preview of your data:")
print(df.head())

```

```python
# Convert rows into text for retrieval
texts = [", ".join([f"{col}: {row[col]}" for col in df.columns]) for _, row in df.iterrows()]

# Use sentence-transformers for embeddings
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Store in FAISS
db = FAISS.from_texts(texts, embeddings)
```

```python
# Add memory so chatbot remembers past queries
memory = ConversationBufferMemory(memory_key="chat_history", return_messages=True)

# Build the retrieval-based Q&A chain
qa_chain = ConversationalRetrievalChain.from_llm(
    llm=llm,
    retriever=db.as_retriever(),
    memory=memory
)
```

```python
query = "Which customer has the highest revenue?"   # Example
answer = qa_chain.run(query)
print("Q:", query)
print("A:", answer)
```

```python
print(qa_chain.run("What are the main columns in the dataset?"))
print(qa_chain.run("Give me a summary of this dataset."))
```

```python
import gradio as gr

def chat_fn(message, history=[]):
    response = qa_chain.run(message)
    history.append((message, response))
    return history, history

demo = gr.ChatInterface(fn=chat_fn, title="CSV Q&A Chatbot (Free, HuggingFace + LangChain)")
demo.launch()
```
